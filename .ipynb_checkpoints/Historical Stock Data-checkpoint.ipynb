```python
import pandas as pd
import tweepy
import time
import json
import random
import requests
import os
import datetime
import time

from config_mine import alphavantage_api_key
```

```python
#setup variables
api_key = alphavantage_api_key
symbol = ""
function = "TIME_SERIES_DAILY"
output_size = "full"
datatype = "json"
time_period = 10 #time period of interest in days, but stocks dont get recorded over weekends

tickers = pd.read_csv(os.path.join("SP_100_Tickers.csv"))
```

```python
#create lists to save data from API requests
ticker_symbol = []
ticker_close = []
ticker_volume = []
ticker_date= []

#iterate over all tickers we imported from the csv
for index, row in tickers.iterrows():
    symbol = row["Symbol"]
    print(f"Symbol: {symbol}")
    
    #try to retrieve the daily stock price information and store the json
    try:
        query_url = "https://www.alphavantage.co/query?" + "function="+function + "&symbol=" + symbol+"&outputsize"+output_size +"&datatype="+datatype+ "&apikey=" + api_key 

        stock_response = requests.get(query_url)

        stock_json = stock_response.json()
        # print historical stock data
        #print(json.dumps(stock_json,indent=4))
    except:
        print(f"Error retrieving {symbol}")
    
    #looks like the API breaks after a few pulls, need to sleep...
    time.sleep(3)
    
    #use the time_period variable to pull back individual day's data out of the json and save in a dataframe
    for i in range(len(stock_json["Time Series (Daily)"])):
        date_of_interest = (datetime.datetime.today()-datetime.timedelta(days=i))
        #check if date weekday, if not get the data
        if date_of_interest.weekday()<5:
            try:
                print(stock_json["Time Series (Daily)"][date_of_interest.strftime('%Y-%m-%d')])
                ticker_close.append(stock_json["Time Series (Daily)"][date_of_interest.strftime('%Y-%m-%d')]["4. close"])
                ticker_volume.append(stock_json["Time Series (Daily)"][date_of_interest.strftime('%Y-%m-%d')]["5. volume"])
                ticker_symbol.append(symbol)
                ticker_date.append(date_of_interest.strftime('%Y-%m-%d'))
            except:
                print(f"Could not print:{date_of_interest}")
                                
```

```python
historical_data = pd.DataFrame({"Ticker":ticker_symbol,
                                "Date":ticker_date,
                               "Close":ticker_close,
                               "Volume":ticker_volume})

historical_data.groupby("Ticker").max()
```
