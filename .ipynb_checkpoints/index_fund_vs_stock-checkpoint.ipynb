```python
import os
import json
import requests
from config import alphavantage_api_key
import pandas as pd
```

```python
top_stocks = pd.read_csv("SP_100_Tickers.csv")
stock_symbol = top_stocks["Symbol"].tolist()

api_key = alphavantage_api_key
#symbol = "VFINX"
function = "TIME_SERIES_MONTHLY"
output_size = "full"
datatype = "json"






query_url = "https://www.alphavantage.co/query?" + "function="+function + "&symbol=" + "AGN"+"&outputsize"+output_size +"&datatype="+datatype+ "&apikey=" + api_key 
monthly_data = requests.get(query_url).json()

print(json.dumps(monthly_data,indent=3))
```

```python
date_list = []
stock_data = []
open_price = []
close_price = []

#https://www.alphavantage.co/query?function=TIME_SERIES_DAILY&symbol=MSFT&outputsize=full&apikey=demo
for symbol in stock_symbol:
    
    query_url = "https://www.alphavantage.co/query?" + "function="+function + "&symbol=" + symbol+"&outputsize"+output_size +"&datatype="+datatype+ "&apikey=" + api_key 
    monthly_data = requests.get(query_url).json()
    
    print(query_url)
    monthly_time_series = monthly_data["Monthly Time Series"]
    
    #gets the date in a list
    for entry in monthly_time_series:
        date_list.append(entry)
           

    #with the date now can access the stock data by date
    for date in date_list:
        try:
            stock_data.append(monthly_data["Monthly Time Series"][date])
        except:
            print("didnt work" + symbol + date)

    #pull the open and close price per date    
    for stuff in stock_data:
        open_price.append(stuff['1. open'])
        close_price.append(stuff['4. close'])
```

```python
# create a dataframe to put together the 3 list 
stock_df = pd.DataFrame({"Date":date_list,
                         "Open Price":open_price,
                         "Close Price":close_price})

#add the symbol
for index,row in stock_df.iterrows():
        stock_df["Symbol"] = symbol
    
   
```

```python
#create list to breakdown date
dates_split = []  
year = []
month = []
day = []

#take date to a list element to split out month and day
stock_date_list = stock_df["Date"].tolist()

#split the month/day/year
for date in stock_date_list:
    dates_split.append(date.split('-' ,2))

#move each spilt into seperate list
for stuff in dates_split:
    year.append(stuff[0])
    month.append(stuff[1])
    day.append(stuff[2])
              
# add month/day/year into dataframe   
stock_df = pd.DataFrame({"Date":date_list,
                         "Open Price":open_price,
                         "Close Price":close_price,
                          "Year" : year,
                         "Month":month,
                          "Day": day})
```

```python
stock_df["Open Price"] = stock_df["Open Price"].astype(float)
stock_df["Close Price"] = stock_df["Close Price"].astype(float)
stock_df["Percent Chg"] = ( stock_df['Close Price'] - stock_df['Open Price']) / stock_df['Open Price']

year_group = stock_df.groupby("Year")
avg_chg_year = pd.DataFrame(year_group["Percent Chg"].mean())
avg_chg_year.head(25)
```
